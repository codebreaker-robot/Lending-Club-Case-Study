```python
#import all the libraries which are required for analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
#load loan csv file
# here we have made index_col=False. It is used to force pandas to not use the first column as the index.
loan=pd.read_csv('D:/Lending Club Case Study/data/loan.csv',index_col=False,error_bad_lines=False,dtype='unicode')
loan.head()
```

```python
#check shape(i.e no of rows and columns)
print("Shape of the loan dataset:")
loan.shape
```

```python
#check datatype of each column
loan.dtypes
```

### Data Cleaning

```python
#find which column having null values
loan.isnull().sum(axis=0)
```

```python
#check how many columns have all missing values
loan.isnull().all(axis=0).sum()
```

```python
#check how many rows have all missing values
loan.isnull().all(axis=1).sum()
```

```python
#drop the columns having null values
loan.drop(loan.iloc[:, 53:105], inplace = True, axis = 1)
```

```python
#drop the columns which are not required for our analysis
loan.drop(columns=['member_id','emp_title','url','collections_12_mths_ex_med','mths_since_last_major_derog','desc','zip_code','mths_since_last_delinq','mths_since_last_record','next_pymnt_d','tot_hi_cred_lim','total_bal_ex_mort','total_bc_limit','total_il_high_credit_limit','tax_liens'],axis =1, inplace=True)
```

```python
#check shape after removing the columns
loan.shape
```

```python
#check top rows after data cleaning
loan.head()
```

```python
# term column contains 'months' at the end of every value,remove 'months'
loan.term=loan.term.str.rstrip('months')
```

```python
#cross verify if the 'months' is removed from term column values or not,check top results
loan.head()
```

```python
#int_rate contains '%' symbol in the values,first need to remove this
loan.int_rate=loan.int_rate.str.rstrip('%')
```

```python
#cross verify if '%' symbol is removed from int_rate or not
loan.head()
```

```python
#In emp_length column there are some 'n/a' values we need fill that or replace it with zero
loan.emp_length.fillna('0',inplace=True)
```

```python
#check top rows from dataframe
loan.head()
```

```python
#As emp_length contains '+','<' symbol and some string like 'years' and 'year',need to clean that which is important for further analysis
loan.emp_length=loan.emp_length.str.rstrip('years ')
loan.emp_length=loan.emp_length.str.rstrip('year ')
```

```python
#cross verify if 'years' and 'year' is removed or not
loan.head()
```

```python
#as we can see from above data that emp_length still contains '+' and '<' symbol,need to fix that issue
#replace '10+' by '10'
loan['emp_length']=loan['emp_length'].apply(lambda x : x.replace('10+','10'))
#cross verify
loan.head()
```

```python
#as emp_length column contains '<1' value,we will round of this value to one specific value.Consider it as 0.5
loan['emp_length']=loan['emp_length'].apply(lambda x : x.replace('< 1','0.5'))
#cross verify if it is rounded off to 0.5 or not
loan.head()
```

```python
#remove'% from revol_util column as it contains '%' symbol
loan.revol_util=loan.revol_util.str.rstrip('%')
```

```python
#convert important columns into numeric which are required for analysis
numeric_cols=['loan_amnt','funded_amnt','int_rate','funded_amnt_inv','installment','annual_inc','dti','emp_length','total_pymnt','total_pymnt_inv']
loan[numeric_cols]=loan[numeric_cols].apply(pd.to_numeric)
```

#### Derived Columns 

```python
#derive columns for our analysis like year and month
# % b-> Returns the first three characters of the month name
# %y ->Returns the year in four-digit format
loan.issue_d = pd.to_datetime(loan.issue_d, format='%b-%y')
loan['issue_year']=loan['issue_d'].dt.year
loan['issue_month']=loan['issue_d'].dt.month
```

#### We have derived  two columns here from issue_d which are issue_ year and issue_month.

```python
#check bottom rows of dataset and cross verify if the year and month column is added or not
loan.tail()
```

### Correlation Matrix

```python
#Plot the correlation for loan dataframe
loan_corr = loan.corr()
sns.clustermap(loan_corr,annot=True,figsize=(10, 9),cmap="YlGnBu")
plt.show()
```

### From the above clustered heatmap,inferences are as follows: 
#### 1.annual_inc is negatively correlated with dti.
#### 2.loan_amnt,funded_amnt and funded_amnt_inv are highly correlated.

```python
#drop the columns which are not required for our analysis
print('Before Dropping of the columns:')
print(loan.shape)
loan.drop(["application_type","policy_code","initial_list_status","pymnt_plan"], axis=1, inplace=True)
print('After Dropping of the columns:')
print(loan.shape)
```

```python
#Analysis of quantitative variable
#Check the quantitative spread of loan_amnt column
loan['loan_amnt'].describe()
```

```python
#Relationship of Funded Amount with defaulters(i.eCharged Off)
#derive new column profit_loss
loan['Profit_Loss']=(loan['total_pymnt']-loan['funded_amnt'])*100/loan['funded_amnt']
#subplot 1
plt.subplot(1,2,1)
plt.title('Defaulters')
sns.boxplot(y=loan[loan.loan_status=='Charged Off'].Profit_Loss)
#subplot 2
plt.subplot(1,2,2)
plt.title('Fully Paid')
sns.boxplot(y=loan[loan.loan_status=='Fully Paid'].Profit_Loss)
plt.show()
```

#### The above plot shows that most of the interest rates are in the range of 10% to 15%

#### Here,most loan grades are A,B and C that means good loans count most.

# Bivariate Analysis

```python
#bivariate analysis of grade vs interest rate and it's boxplot
plt.figure(figsize=(10,6))
sns.barplot(x='grade',y='annual_inc',hue="loan_status",data=loan,order='ABCDEFG',estimator=np.mean)
plt.title('Grade Vs Annual Income')
plt.xlabel('Grade')
plt.ylabel('Annual Income')
plt.show()
```

### Conclusion: 
#### 1.The people who paid fully having almost similar interest range for all grades.
#### 2.Charged Off(defaulters) having lower annual income.

```python
#bivariate analysis of grade vs interest rate and it's boxplot
plt.figure(figsize=(10,6))
sns.boxplot(y='int_rate', x='grade',data=loan,order = 'ABCDEFG')
plt.xlabel('Grade')
plt.title('Grade vs Interest Rate')
plt.ylabel('Interest Rate')
plt.show()
```

### Conclusion:
####  1.Interest rate is increasing with grades from A to F
####  2.Higher the grade,lower the interest rate.

```python
#bivariate analysis of term vs interest rate and it's boxplot
plt.figure(figsize=(10,6))
sns.boxplot(y='int_rate',x='term',data=loan)
plt.title('Term Vs Interest Rate')
plt.xlabel('Term')
plt.ylabel('Interest Rate')
plt.show()
```

### Conclusion:
####  Avg interest rate is higher for 60 months loan term.

```python
def_count = loan[loan['loan_status']=='Charged Off'].groupby('addr_state')['loan_status'].count().reset_index()
plt.figure(figsize=(12,10))
sns.barplot(y='addr_state', x='loan_status',data=def_count)
plt.xlabel('Defaulter Count')
plt.ylabel('State')
plt.show()
```

### Conclusion:
#### 1.From the above plot,we can say that most defaulters from  state CA.
#### 2.FL,NY,TX having good number of charged off/defaulters.

```python
#bivariate analysis of issue_year and interest rate
plt.figure(figsize=(10,6))
sns.boxplot(y='int_rate', x='issue_year',data=loan)
plt.title('Issue Year vs Interest Rate')
plt.xlabel('Loan Issue Year')
plt.ylabel('Interest Rate')
plt.show()
```

### Conclusion:
#### With increase in year,interest rate is also increasing slowly.

### Multivariate Analysis 

```python
#select main columns from dataset which are important for our analysis
multi_loan = loan[['loan_amnt', 'annual_inc', 'issue_year','int_rate','loan_status']].copy()
multi_loan.head()
```

### Conclusion:
